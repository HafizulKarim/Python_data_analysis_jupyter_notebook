```python
#importing necessary python libraries for analysis.
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#loading desired data set into notebook.
df = pd.read_csv('heart.csv')
df.head(10)
```

```python
#checking how many columns are availabe.
df.columns
```

```python
#checking is there any null values in the columns.
df.[age']
```

```python
#informations about the data set.
df.info()
```

```python
#Histogram of above columns available in data frame.
df.hist(bins=50, grid=False, figsize = (20,15));
```

```python
# a descriptive statestics of the data frame.
df.describe()
```

```python
# Finding how many people have heart disease and how many people doesn't have heart disease?
df.target.value_counts()
# where '1' represent heart disease and '0' represent doesn't have.
```

```python
# we can see above result via a bar graph.
df.target.value_counts().plot(kind = 'bar', color = ['salmon','gray'])
plt.title('Heart Disease Values')
plt.xlabel('1 = Disease, 0 = No Disease')
plt.ylabel('Amount');
```

```python
# Now we will see which gender has most heart disease.
df.sex.value_counts()
```

```python
# we can see the result through a pie chart.
df.sex.value_counts().plot(kind = 'pie', figsize = (8,6))
plt.legend(['Male','Female'])
```

```python
# have heart disease or haven't heart disease corresponding with gender?
pd.crosstab(df.sex,df.target)
```

```python
# we might see the result through plotting.
sns.countplot(x = 'target', data = df, hue = 'sex')
```

```python
# people of which sex has which type(0 1 2 3) of chest pain(cp) most?
df.cp.value_counts()
```

```python
#plotting a bar chart of above result.
df.cp.value_counts().plot(kind = 'bar', color = ['salmon','gray','green','orange'] )
plt.title("Chest pain vs number of people");
```

```python
pd.crosstab(df.sex,df.cp)
```

```python
pd.crosstab(df.sex,df.cp).plot(kind = 'bar', color = ['gray','orange','green','orange'] )
plt.title("Types of chest pain w.r.t sex")
plt.xlabel('0 = Female, 1 = Male');
```

```python
# people with which chest pain are most pron to have heart disease?
pd.crosstab(df.cp,df.target)
```

```python
# plotting a bar chart of above result.
sns.countplot(x = 'cp', data = df, hue = 'target')
```

```python
# we will see age vs heart rate when people have and haven't heart disease.
# we observe it through a distribution plot.
sns.displot(x = 'age', data= df, bins = 30, kde = True);
```

```python
# We also can see a scatter plot for above result.
# Here thalach is represent maximum heart rate column.
plt.figure(figsize = (10, 6))
plt.scatter(df.age[df.target == 1],df.thalach[df.target == 1],c = 'blue')
plt.scatter(df.age[df.target == 0],df.thalach[df.target == 0],c = 'red')
plt.title("Max heart rate vs age") 
plt.xlabel('Age')
plt.ylabel('Maximum heart rate')
plt.legend(['Disease','No Diseae']);
```

```python
# we will see age vs serum cholesteral when people have and haven't heart disease.
# we observe it through a distribution plot.
plt.figure(figsize = (10, 6))
plt.scatter(df.age[df.target == 1],df.chol[df.target == 1],c = 'gray')
plt.scatter(df.age[df.target == 0],df.chol[df.target == 0],c = 'salmon')
plt.title("Serum cholesteral vs age") 
plt.xlabel('Age')
plt.ylabel('Serum cholesteral')
plt.legend(['Disease','No Diseae']);
```
